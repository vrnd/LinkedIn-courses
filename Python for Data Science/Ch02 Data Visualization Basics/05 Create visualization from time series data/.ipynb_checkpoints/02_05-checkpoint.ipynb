### The simplest time series plot

I got an UTF-8 issue, so to check what encoding the csv is in,
It is best to put in a large number such as 100000, smaller sample sizes maynot be accurate in some cases

```python
import chardet
with open('Superstore-Sales.csv','rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
    
print(result)
```

```python
address = 'Superstore-Sales.csv'
df = pd.read_csv(address, index_col='Order Date', parse_dates=True, encoding = 'Windows-1252')
df.head().T
```

```python
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
```
